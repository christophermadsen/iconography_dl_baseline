```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('seaborn')
```

```python
val1 = np.genfromtxt('best_vgg_1st_strategy/vgg_training_metrics_1st_strategy.csv', delimiter=',')
val2 = np.genfromtxt('best_vgg_2nd_strategy/vgg_training_metrics.csv', delimiter=',')
val3 = np.genfromtxt('best_vgg_3rd_strategy/vgg_training_metrics.csv', delimiter=',')
```

```python
plt.figure()
plt.plot(range(1, 37), val1[0])
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')
plt.savefig('evaluation_files/graphs/val_loss_strat1.jpg')
plt.show()
```

```python
plt.figure()
plt.plot(range(1, 201), val2[0], color='teal')
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')
plt.savefig('evaluation_files/graphs/val_loss_strat2.jpg')
plt.show()
```

```python
plt.figure()
plt.plot(range(1, 37), val1[1])
plt.xlabel('Epoch')
plt.ylabel('Accurracy')
plt.savefig('evaluation_files/graphs/val_acc_strat1.jpg')
plt.show()
```

```python
plt.figure()
plt.plot(range(1, 201), val2[1], color='teal')
plt.xlabel('Epoch')
plt.ylabel('Accurracy')
plt.savefig('evaluation_files/graphs/val_acc_strat2.jpg')
plt.show()
```

```python
x1, x2 = range(1, 37), range(37, 70)
```

```python
plt.figure()
plt.plot(x1, val1[0])
plt.plot(x2, val3[0])
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')
plt.savefig('evaluation_files/graphs/val_loss_strat3.jpg')
plt.show()
```

```python
plt.figure()
plt.plot(x1, val1[1])
plt.plot(x2, val3[1])
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.savefig('evaluation_files/graphs/val_acc_strat3.jpg')
plt.show()
```

```python
plt.figure()
x = []
csum = 0
N = len(val1[1])
for v in val1[1]:
    csum += v
    x.append(csum/N)

plt.plot(x)
plt.xlabel('Epoch')
plt.ylabel('Accumulative Mean Accuracy')
plt.savefig('evaluation_files/graphs/ama_strat1.jpg')
plt.show()
```

```python
plt.figure()
x = []
csum = 0
N = len(val2[1])
for v in val2[1]:
    csum += v
    x.append(csum/N)

plt.plot(x, color='teal')
plt.xlabel('Epoch')
plt.ylabel('Accumulative Mean Accuracy')
plt.savefig('evaluation_files/graphs/ama_strat2.jpg')
plt.show()
```

```python
plt.figure()
x = []
csum = 0
data = np.concatenate((val1[1], val3[1]))
N = len(val1[1]) + len(val3[1])
for v in data:
    csum += v
    x.append(csum/N)

plt.plot(0, 0)
plt.plot(x)
plt.xlabel('Epoch')
plt.ylabel('Accumulative Mean Accuracy')
plt.savefig('evaluation_files/graphs/ama_strat3.jpg')
plt.show()
```
