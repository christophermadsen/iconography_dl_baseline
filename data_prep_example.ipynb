```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as F
import PIL
from torchvision import transforms
from preprocessing.squarepad import SquarePad
```

```python
transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),
        SquarePad(),
        transforms.Resize((224, 224)),
        transforms.transforms.ToPILImage()
    ])
```

```python
img_path = 'test-data/seb.jpg'
img = PIL.Image.open(img_path)
img
```

```python
t_img = transforms.ToTensor()(img)
t_img = transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))(t_img)
s_img = transforms.transforms.ToPILImage()(t_img)
# s_img.save('prep1.jpg')
s_img
```

```python
t_img = SquarePad()(t_img)
s_img = transforms.transforms.ToPILImage()(t_img)
# s_img.save('prep2.jpg')
s_img
```

```python
t_img = transforms.Resize((224, 224))(t_img)
s_img = transforms.transforms.ToPILImage()(t_img)
# s_img.save('prep3.jpg')
s_img
```
